```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data
file_path = 'results_70b.csv'
data = pd.read_csv(file_path)
```

```python
# Filter data for fixed_5, semantic_5, and recursive_5
methods_to_compare = ['fixed_5', 'recursive_5','semantic_5']
filtered_data = data[data['method'].isin(methods_to_compare)]

# Metrics to plot
metrics = ['Faithfulness', 'Relevancy', 'FactualCorrectness', 'Overall']

# Create bar plot for side-by-side comparison of chunking methods
x = range(len(metrics))  # Positions for metrics on x-axis
width = 0.25  # Width of each bar

plt.figure(figsize=(12, 8))

# Plot bars for each method
for idx, method in enumerate(methods_to_compare):
    plt.bar(
        [pos + idx * width for pos in x],  # Position adjustment
        filtered_data[filtered_data['method'] == method][metrics].values[0],  # Metric values
        width=width,
        label=method.split('_')[0].capitalize()  # Simplify method names (Fixed, Semantic, Recursive)
    )

# Add labels and title
plt.title('Comparison of Chunking Methods')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.xticks([pos + width for pos in x], metrics)
plt.legend(title='Methods')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python

# Filter data for semantic chunk sizes
semantic_methods = ['semantic_3', 'semantic_5', 'semantic_8', 'semantic_12']
semantic_data = data[data['method'].isin(semantic_methods)]

# Extract metrics and prepare data for line plot
semantic_plot_data = semantic_data.set_index('method')[metrics]

# Plotting line plot for chunk sizes
chunk_sizes = ['3', '5', '8', '12']
plt.figure(figsize=(12, 8))
for metric in metrics:
    plt.plot(chunk_sizes, semantic_plot_data[metric], marker='o', label=metric)

plt.title('Comparison of Chunk Sizes for Semantic Chunking')
plt.xlabel('Chunk Size')
plt.ylabel('Scores')
plt.grid(True)
plt.legend(title='Metrics')
plt.tight_layout()
plt.show()
    
```

```python
# Filter data for the specified methods
methods_to_compare = ['semantic_8', 'semantic_hyde_8', 'semantic_query2doc_8', 'semantic_hybrid_8']
filtered_data = data[data['method'].isin(methods_to_compare)]

# Rename methods for clarity
filtered_data['method'] = filtered_data['method'].replace({
    'semantic_8': 'No Method',
    'semantic_hyde_8': 'HyDE',
    'semantic_query2doc_8': 'Query2Doc',
    'semantic_hybrid_8': 'Hybrid'
})

# Metrics to plot
metrics = ['Faithfulness', 'Relevancy', 'FactualCorrectness', 'Overall']

# Create bar plot for side-by-side comparison
x = range(len(metrics))  # Positions for metrics on x-axis
width = 0.2  # Width of each bar

plt.figure(figsize=(12, 8))

# Plot bars for each method
for idx, method in enumerate(filtered_data['method'].unique()):
    plt.bar(
        [pos + idx * width for pos in x],  # Position adjustment
        filtered_data[filtered_data['method'] == method][metrics].values[0],  # Metric values
        width=width,
        label=method
    )

# Add labels and title
plt.title('Comparison of Semantic Methods (HyDE, Query2Doc, Hybrid) with No Method')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.xticks([pos + width for pos in x], metrics)
plt.legend(title='Methods')
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Filter data for reranking methods
methods_to_compare = ['semantic_8', 'semantic_bge_8', 'semantic_tilde_8']
filtered_data = data[data['method'].isin(methods_to_compare)]

# Rename methods for clarity
filtered_data['method'] = filtered_data['method'].replace({
    'semantic_8': 'No Rerank',
    'semantic_bge_8': 'BGE',
    'semantic_tilde_8': 'TILDE'
})

# Metrics to plot
metrics = ['Faithfulness', 'Relevancy', 'FactualCorrectness', 'Overall']

# Create bar plot for side-by-side comparison
x = range(len(metrics))  # Positions for metrics on x-axis
width = 0.25  # Width of each bar

plt.figure(figsize=(12, 8))

# Plot bars for each method
for idx, method in enumerate(filtered_data['method'].unique()):
    plt.bar(
        [pos + idx * width for pos in x],  # Position adjustment
        filtered_data[filtered_data['method'] == method][metrics].values[0],  # Metric values
        width=width,
        label=method  # Legend label
    )

# Add scores on bars
for idx, method in enumerate(filtered_data['method'].unique()):
    for i, value in enumerate(filtered_data[filtered_data['method'] == method][metrics].values[0]):
        plt.text(i + idx * width, value + 0.01, f'{value:.2f}', ha='center')

# Add labels and title
plt.title('Comparison of Reranking Methods')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.xticks([pos + width for pos in x], metrics)
plt.legend(title='Methods', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Filter data for summarization vs no summarization methods
methods_to_compare = ['semantic_tilde_8', 'semantic_tilde_summarize_8']
filtered_data = data[data['method'].isin(methods_to_compare)]

# Rename methods for clarity
filtered_data['method'] = filtered_data['method'].replace({
    'semantic_tilde_8': 'No Summarization',
    'semantic_tilde_summarize_8': 'Summarization'
})

# Metrics to plot
metrics = ['Faithfulness', 'Relevancy', 'FactualCorrectness', 'Overall']

# Create bar plot for side-by-side comparison
x = range(len(metrics))  # Positions for metrics on x-axis
width = 0.35  # Width of each bar

plt.figure(figsize=(12, 8))

# Plot bars for each method
for idx, method in enumerate(filtered_data['method'].unique()):
    plt.bar(
        [pos + idx * width for pos in x],  # Position adjustment
        filtered_data[filtered_data['method'] == method][metrics].values[0],  # Metric values
        width=width,
        label=method  # Legend label
    )

# Add scores on bars
for idx, method in enumerate(filtered_data['method'].unique()):
    for i, value in enumerate(filtered_data[filtered_data['method'] == method][metrics].values[0]):
        plt.text(i + idx * width, value + 0.01, f'{value:.2f}', ha='center')

# Add labels and title
plt.title('Comparison of Summarization vs No Summarization')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.xticks([pos + width / 2 for pos in x], metrics)

# Position legend outside the plot
plt.legend(title='Methods', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Show plot
plt.show()
```

```python

# Adjust 'base' method's overall score as half of Relevancy and FactualCorrectness
rag_comparison = data[data['method'].isin(['base', 'semantic_tilde_summarize_8'])].copy()
rag_comparison['NewOverall'] = (rag_comparison['Relevancy'] + rag_comparison['FactualCorrectness']) / 2

# Prepare data for plotting
metrics_final = ['Relevancy', 'FactualCorrectness', 'NewOverall']
rag_comparison_plot = rag_comparison.set_index('method')[metrics_final]

# Rename and plot
rag_comparison_plot = rag_comparison_plot.rename(columns={'NewOverall': 'Overall'})
rag_comparison_plot = rag_comparison_plot.rename(index={'base': 'No RAG', 'semantic_tilde_summarize_8': 'RAG'}).transpose()

plt.figure(figsize=(12, 8))
colors = ['#1f77b4', '#ff7f0e']
ax = rag_comparison_plot.plot(kind='bar', width=0.6, figsize=(12, 6), color=colors)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge', padding=3)

plt.title('RAG vs No RAG')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.xticks(rotation=0)
plt.legend(title='Methods', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
    
```
